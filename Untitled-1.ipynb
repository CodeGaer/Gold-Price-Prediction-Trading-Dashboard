```python
corr = df.corr(numeric_only=True)

corr['close'].sort_values(ascending=False)  
```

```python
df['lag_1'] = df['close'].shift(1)
df['lag_2'] = df['close'].shift(2)
df['lag_3'] = df['close'].shift(3)
df['lag_7'] = df['close'].shift(7)

# Predict return
df['target'] = (df['close'].shift(-1) > df['close']).astype(int)

df = df.dropna()


```

```python
features = [
    'lag_1','lag_2','lag_3','lag_7',
    'rsi','volatility_7','volatility_30',
    'macd','macd_signal'
]

X = df[features]
y = df['target']
```

```python
split = int(len(df) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=300,
    max_depth=5,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```

```python
df['ma_10'] = df['close'].rolling(10).mean()
df['ma_20'] = df['close'].rolling(20).mean()
df['ma_50'] = df['close'].rolling(50).mean()
```

```python
df['price_ma20_diff'] = df['close'] - df['ma_20']
df['price_ma50_diff'] = df['close'] - df['ma_50']
```

```python
df['mom_5']  = df['close']/df['close'].shift(5)  - 1
df['mom_10'] = df['close']/df['close'].shift(10) - 1
df['mom_20'] = df['close']/df['close'].shift(20) - 1
```

```python
df['vol_14'] = df['close'].pct_change().rolling(14).std()
df['vol_30'] = df['close'].pct_change().rolling(30).std()
```

```python
df['high_low_range'] = (df['high'] - df['low']) / df['close']
```

```python
df['trend_20'] = (df['ma_10'] > df['ma_20']).astype(int)
df['trend_50'] = (df['ma_20'] > df['ma_50']).astype(int)
```

```python
df['ml_prediction'] = np.nan
df.loc[X_test.index, 'ml_prediction'] = y_pred
```

```python
df['signal'] = (df['ml_prediction'] > 0.5).astype(int)
```

```python
df['daily_return'] = df['close'].pct_change()
```

```python
df['strategy_return'] = df['signal'].shift(1) * df['daily_return']
```

```python
df = df.dropna()
```
